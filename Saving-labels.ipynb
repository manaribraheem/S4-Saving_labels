```python
"""1- Function that returns labels and save every label over the original image in the their folder (not recommended!)"""
#Don't apply this function. Juts try to understand what it does.

def save_labels1(dir_path):
    
    for item in df4ToList_no_dup:  #iterate over items in the list of filename
        
        abs_path = os.path.join(dir_path, item) #create path
        
        im = Image.open(abs_path)# open an image    
        
        fig, ax = plt.subplots() # create an ax and figure (according to this ax the rectangle will be positionned)  
        
        ax.imshow(im) #show the image: till now no rectangles  
        
        for idx, row in df4.iterrows():# interate over row in the dataframe dfc
            
            if item == row["filename"]: # if the item = the filename at this row
                
                rect = patches.Rectangle((row ["X"],row["Y"]),row["Width"],row["Height"], linewidth=1.1, edgecolor='r', facecolor='none')#draw a rectangle using X, Y for position and width ang height for dimensions
                
                ax.add_patch(rect) # add every rectangle drawn in the second boucleto the image we opened in the first boucle.
                
                plt.savefig(abs_path)

```
